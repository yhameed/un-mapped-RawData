```python
#import necessary libraries
import pandas as pd
from pandas import DataFrame
import numpy
import matplotlib.pyplot as plt
import glob
import os
```

```python
#open files and check dataframes
df1= pd.read_csv('dstrCAW_1.csv')
df2= pd.read_csv('dstrCAW_2013.csv')
df3= pd.read_csv('dstrCAW_1_2014.csv')
df1

```

```python
pd.set_option('display.max_columns', None)
df2
```

```python
#taking out fields that cannot be joined with data from previous years
df4= df3[['States/UTs','District', 'Year','Rape','Kidnapping & Abduction_Total','Dowry Deaths','Assault on Women with intent to outrage her Modesty_Total','Insult to the Modesty of Women_Total','Cruelty by Husband or his Relatives','Importation of Girls from Foreign Country']]
df4
```

```python
#renaming columns to concatonate data frames
df4 = df4.rename(columns = {
    'States/UTs':'STATE/UT',
    'District':'DISTRICT',
    'Year':'Year',
    'Rape':'Rape',
    'Kidnapping & Abduction_Total':'Kidnapping and Abduction',
    'Dowry Deaths':'Dowry Deaths',
    'Assault on Women with intent to outrage her Modesty_Total':'Assault on women with intent to outrage her modesty',
    'Insult to the Modesty of Women_Total':'Insult to modesty of Women',
    'Cruelty by Husband or his Relatives':'Cruelty by Husband or his Relatives',
    'Importation of Girls from Foreign Country':'Importation of Girls',
})
#concatonating data from previous years
frames = [df1, df2, df4]
df = pd.concat(frames)
df.columns = map(str.lower, df.columns) # lowercase headers
df= df.rename(columns = { #remove slash from variable name, and rename to more suitable and shorter names
    'state/ut': 'state_ut',
    'kidnapping and abduction':'kidnapping_abduction',
    'dowry deaths':'dowry_deaths',
    'assault on women with intent to outrage her modesty' : 'assault_intnt_outrage_modesty',
    'insult to modesty of women': 'inslt_modesty_women',
    'cruelty by husband or his relatives': 'cruelty_husban_reltvs',
    'importation of girls':'in_trafficking'})
df.state_ut = df.state_ut.str.lower()
df.district = df.district.str.lower()
df
```

```python
# dropping records with totals
df = df[(df.district != 'total')]
df = df[(df.district != 'delhi ut total')]
df = df[(df.district != 'zz total')]
df
```

```python
# renaming inconsistant state names
df = df.replace('delhi ut', 'delhi')
df = df.replace ('a & n islands', 'a&n islands')
df = df.replace ('d & n haveli', 'd&n haveli')
df = df.replace ('d & n haveli', 'd&n haveli')
df.year = df.year.astype(str)
#there are inconsistant names in the districts as well but im going to skip those for now
df
```
